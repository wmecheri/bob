```python
from collections import Counter
from datasets import load_from_disk
import matplotlib.pyplot as plt
import numpy as np
```

```python
balanced_path = "C:/Users/wassi/Documents/Fanshawe/S2/pytorch/Project/wikiart_classification/data/balanced_dataset"

balanced_dataset = load_from_disk(balanced_path)
print("Balanced dataset loaded from disk.")
print(f"Dataset size: {len(balanced_dataset)} images.")

unique_styles = sorted(set(balanced_dataset["style"]))
print("Unique style labels in the balanced dataset:", unique_styles)

style_mapping = {orig: idx for idx, orig in enumerate(unique_styles)}
print("Style mapping:", style_mapping)

def remap_style_labels(example):
    example["style"] = style_mapping[example["style"]]
    return example

remapped_dataset = balanced_dataset.map(remap_style_labels)
print("Style labels remapped.")
from pathlib import Path

remapped_path = "C:/Users/wassi/Documents/Fanshawe/S2/pytorch/Project/wikiart_classification/data/remapped_dataset"
remapped_dataset.save_to_disk(remapped_path)
print("Remaped dataset saved to disk.")
```

```python
def plot_distribution(dataset, column, title):
    counts = Counter(dataset[column])
    labels = sorted(counts.keys())
    values = [counts[label] for label in labels]
    
    plt.figure(figsize=(8, 4))
    plt.bar(labels, values, color='skyblue')
    plt.xlabel(column.capitalize())
    plt.ylabel("Count")
    plt.title(title)
    plt.show()

plot_distribution(remapped_dataset, "genre", "Balanced Dataset - Genre Distribution")
plot_distribution(remapped_dataset, "style", "Balanced Dataset - Style Distribution")
```
