```python
import random
import json
import gc
from collections import defaultdict, Counter
from datasets import load_from_disk, load_dataset
import torchvision.transforms as transforms
from torch.utils.data import DataLoader
from torch.utils.data._utils.collate import default_collate
import matplotlib.pyplot as plt
import numpy as np
```

```python
RANDOM_SEED = 42
random.seed(RANDOM_SEED)

def load_full_dataset():
    ds = load_dataset("huggan/wikiart")["train"]
    pair_to_indices = defaultdict(list)
    for idx, (g, s) in enumerate(zip(ds['genre'], ds['style'])):
        pair_to_indices[(g, s)].append(idx)

    genres = sorted(set(ds['genre']))
    styles = sorted(set(ds['style']))
    return ds, pair_to_indices, genres, styles

full_ds, full_pair_to_indices, full_genres, full_styles = load_full_dataset()
print("Full dataset loaded.")
print(f"Number of genres in full dataset: {len(full_genres)}")
print(f"Number of styles in full dataset: {len(full_styles)}")
```

```python
RANDOM_SEED = 42
random.seed(RANDOM_SEED)

def compute_available(pair_to_indices, genres, styles):
    available = {}
    for g in genres:
        for s in styles:
            available[(g, s)] = len(pair_to_indices[(g, s)])
    return available

def filter_styles(available, genres, styles, style_threshold):
    style_total = {s: sum(available[(g, s)] for g in genres) for s in styles}
    retained_styles = [s for s in styles if style_total[s] >= style_threshold]
    return sorted(retained_styles)

def feasible(k, genres, styles, available, num_genres):
    T_genre = len(styles) * k
    T_style = num_genres * k
    total_required = num_genres * T_genre
    import networkx as nx
    G_flow = nx.DiGraph()
    source, sink = "src", "sink"
    for g in genres:
        G_flow.add_edge(source, f"genre_{g}", capacity=T_genre)
    for s in styles:
        G_flow.add_edge(f"style_{s}", sink, capacity=T_style)
    for g in genres:
        for s in styles:
            cap = available.get((g, s), 0)
            if cap > 0:
                G_flow.add_edge(f"genre_{g}", f"style_{s}", capacity=cap)
    flow_value, flow_dict = nx.maximum_flow(G_flow, source, sink)
    return flow_value == total_required, flow_dict

def find_max_k(genres, styles, available, num_genres):
    max_k = 0
    k = 1
    while True:
        is_feasible, _ = feasible(k, genres, styles, available, num_genres)
        if is_feasible:
            max_k = k
            k += 1
        else:
            break
    return max_k

def get_baseline_assignment(k, genres, styles, available, num_genres):
    _, flow_dict = feasible(k, genres, styles, available, num_genres)
    assignment = {}
    for g in genres:
        for s in styles:
            count = flow_dict.get(f"genre_{g}", {}).get(f"style_{s}", 0)
            if count > 0:
                assignment[(g, s)] = count
    return assignment

def sample_final_indices(assignment, pair_to_indices):
    final_indices = []
    for (g, s), count in assignment.items():
        indices = pair_to_indices[(g, s)]
        if len(indices) < count:
            raise ValueError(f"Not enough samples for genre {g} and style {s}.")
        final_indices.extend(random.sample(indices, count))
    return final_indices

def create_balanced_dataset(ds, pair_to_indices, genres, styles):
    available = compute_available(pair_to_indices, genres, styles)
    style_threshold = 2400
    styles_filtered = filter_styles(available, genres, styles, style_threshold)
    num_styles = len(styles_filtered)
    num_genres = len(genres)
    max_k = find_max_k(genres, styles_filtered, available, num_genres)
    if max_k == 0:
        raise ValueError("No balanced subset exists with the given constraints.")
    baseline_assignment = get_baseline_assignment(max_k, genres, styles_filtered, available, num_genres)
    final_indices = sample_final_indices(baseline_assignment, pair_to_indices)
    balanced_ds = ds.select(final_indices)
    return balanced_ds

# Create balanced dataset.
balanced_dataset = create_balanced_dataset(full_ds, full_pair_to_indices, full_genres, full_styles)
print(f"Balanced dataset has {len(balanced_dataset)} images.")

balanced_dataset.save_to_disk("C:/Users/wassi/Documents/Fanshawe/S2/pytorch/Project/wikiart_classification/data/balanced_dataset")
print("Balanced dataset saved to disk.")
```

```python
balanced_dataset = load_from_disk("C:/Users/wassi/Documents/Fanshawe/S2/pytorch/Project/wikiart_classification/data/balanced_dataset")

def plot_distribution(dataset, column, title):
    counts = Counter(dataset[column])
    labels = sorted(counts.keys())
    values = [counts[label] for label in labels]
    
    plt.figure(figsize=(8, 4))
    plt.bar(labels, values, color='skyblue')
    plt.xlabel(column.capitalize())
    plt.ylabel("Count")
    plt.title(title)
    plt.show()

plot_distribution(balanced_dataset, "genre", "Balanced Dataset - Genre Distribution")
plot_distribution(balanced_dataset, "style", "Balanced Dataset - Style Distribution")
```
